# Custom Color Cycle

```python
import itertools

import matplotlib.pyplot as plt
import numpy as np
```

## Cylce Standard Colors

```python
a = np.arange(0, 5*np.pi, .1)

for i in range(10):
    plt.plot(np.sin(a)*(i-1.5), label=i)
```

## Cycle through list of custom colors

```python
a = np.arange(0, 5*np.pi, .1)

my_colors = ['r', 'b', 'k']

for i, color in zip(range(10), itertools.cycle(my_colors)):
    plt.plot(np.sin(a)*(i-1.5), label=i, color=color)
```
