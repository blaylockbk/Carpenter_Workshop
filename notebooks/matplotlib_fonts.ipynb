```python
# fonts = [
#     'monospace',
#     'Nimbus Sans',
#     'Nimbus Sans Narrow',
#     'DejaVu Sans',
#     'Fira Code',
#     "Hubot-Sans",
#     "Mona-Sans",
# ]

# Get all the available fonts in Matplotlib
fonts = sorted([f.name for f in matplotlib.font_manager.fontManager.ttflist])
fonts = np.unique(fonts)
remove_these = [
    "cmex10",
    "STIXGeneral",
    "STIXNonUnicode",
    "STIXSizeFiveSym",
    "STIXSizeFourSym",
    "STIXSizeOneSym",
    "STIXSizeThreeSym",
    "STIXSizeTwoSym",
    "OpenSymbol",
    'DejaVu Serif Display',
    'DejaVu Sans Display',
    "D050000L"
]
mask = ~np.isin(fonts, remove_these)
fonts = fonts[mask]

for i, font in enumerate(fonts):
    plt.text(
        0, i / 10, f"{font} ", fontsize=15, ha="right", alpha=0.5, fontweight="bold"
    )
    plt.text(0, i / 10, f" {font}", fontfamily=font, fontsize=15)
    plt.text(
        1, i / 10, f"The Quick Brown Fox", fontfamily=font, fontsize=15, ha="right"
    )


plt.axis("off")
```

```python
i = 0
for fontweight in [100, 'normal', 'bold', 800]:
    for stretch in [100, 'normal', 800]:
        style = dict(
            fontweight=fontweight, stretch=stretch, fontsize=15,
        )

        plt.text(
            0, i / 10, f"weight={fontweight}, stretch={stretch}", ha="right", alpha=0.5
        )
        plt.text(0, i / 10, f" Mona-Sans", fontfamily="Mona-Sans", **style)
        plt.text(
            1, i / 10, f"The Quick Brown Fox", fontfamily="Mona-Sans", ha="right", **style
        )

        i += 1

plt.axis("off")
```
