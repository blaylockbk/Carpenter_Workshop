# Wind Power Law 
The wind power law is used to esimate the wind at a different height above ground.

https://en.wikipedia.org/wiki/Wind_profile_power_law

```python
from toolbox.wind import wind_profile_power_law
from paint.simple_colors import simple_colors

import numpy as np
import matplotlib.pyplot as plt

simple_colors().set_rcParams
```

```python
plt.figure(figsize=[4,5], dpi=150)

new_heights = np.arange(0,30.1,.1)
_speeds = range(4,12,2)

for i in _speeds:
    # For the speed i at 10 m height, what is the speed at all other heights?
    adjusted_wind_speed = wind_profile_power_law(i, 10, new_heights)
    
    plt.fill_between(new_heights, adjusted_wind_speed, np.ones_like(new_heights)*i, color='k', alpha=.1, lw=.5, zorder=2)
    plt.plot(new_heights, adjusted_wind_speed, label=f"{i}" + " $m s^{-1}$", zorder=2, lw=2.5)
    plt.scatter(10, i, zorder=2, lw=.5, ec='k', s=70)

# Cosmetics

plt.ylabel("Adjusted Wind Speed ($m s^{-1}$)")
plt.xlabel("Height (m)")

plt.grid(lw=.5, ls='--', alpha=.2, zorder=1)

plt.xlim(new_heights[0], new_heights[-1])
plt.ylim(ymin=2)
plt.legend(title='Wind at 10 m', fontsize=8, ncol=1, loc=1, framealpha=.95)#, bbox_to_anchor=(1, 1))
```
