```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patheffects as path_effects

from paint.palettes import sesameStreet, NASA, muppet, university_of_utah, naval_research_lab, BYU, usps, NOAA, bootstrap
```

```python
def show_palette(name):
    plt.figure(figsize=[3,len(name)])

    for i, (key, value) in enumerate(name.items()):

        text = plt.text(0, -i/len(name), f"{key.upper():^20}", color='w', alpha=.5, 
                 va='center',
                 fontsize=20, fontweight='bold', fontfamily='monospace',
                 bbox=dict(facecolor=value, edgecolor='k', lw=2, boxstyle='Round'))

        text.set_path_effects([
            path_effects.Stroke(linewidth=1, foreground='black'), 
            path_effects.Normal(),
            #path_effects.SimpleLineShadow(linewidth=10),
            #path_effects.Normal()
            ])
        
        plt.scatter(0, -i/len(name), color=value, s=2000, edgecolor='k', lw=2, zorder=100, clip_on=False)

    plt.gca().set_axis_off()
```

```python
show_palette(naval_research_lab)
```

```python
show_palette(NOAA)
```

```python
show_palette(usps)
```

```python
show_palette(bootstrap)
```
