````python
import matplotlib.pyplot as plt

from paint.simple_colors import simple_colors
````

````python
simple_colors().color_list
````

````python
# Here are matplotlib's default color cycler...
ax = plt.axes()
for i in range(20):
    ax.scatter(i,i, s=250)
    
plt.title("Default Matplotlib Color Cycle")
````

````python
# Change the color cycle to simple_colors by setting the rcParams
simple_colors().set_rcParams

ax = plt.axes()
for i in range(22):
    ax.scatter(i,i, s=250, ec='k', lw=.1)
plt.title("Simple Colors: 95% accessible (default)")
````

````python
# Change to 100% accessibility
simple_colors(1).set_rcParams

ax = plt.axes()
for i in range(20):
    ax.scatter(i,i, s=250, ec='k', lw=.1)

plt.title("Simple Colors: 100% accessible")
````

````python
# Change to 99% accessibility
simple_colors(.99).set_rcParams

ax = plt.axes()
for i in range(20):
    ax.scatter(i,i, s=250, ec='k', lw=.1)
plt.title("Simple Colors: 99% accessible")
````

````python
# Change to 99.99% accessibility
simple_colors(.9999).set_rcParams

ax = plt.axes()
for i in range(20):
    ax.scatter(i,i, s=250, ec='k', lw=.1)
    
plt.title("Simple Colors: 99.99% accessible")
````
